```python
# finding the missing values
df.isnull().sum()
```

```python
# sum of Cases
Total_cases = df.Cases.sum()
Total_cases
```

```python
# sum of Deaths
Total_deaths = df.Deaths.sum()
Total_deaths
```

```python
# sum of Recovered
Total_recovered = df.Recovered.sum()
Total_recovered
```

```python
# calculating death rate
death_rate = Total_deaths/Total_cases * 100
death_rate
```

```python
# calculating recovered rate
recovered_rate = Total_recovered/Total_cases * 100
recovered_rate
```

```python
#visualize the nan values in heatmap
sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
#What is the mode of "Travel_history"
print('mode is:', df['Travel_history'].mode()[0])
```

```python
# filling the missing values of 'Travel_History' column
df.Travel_history.fillna(df['Travel_history'].mode()[0], inplace=True)
```

```python
# Summarize
df.describe() #summary of statistics
```

```python
#visualize & explore
sns.countplot(data=df, x='Province')
plt.xticks(rotation=90)
```

```python
#pairplot
sns.pairplot(data=df, hue='Province')
```

```python
# pairplot
sns.pairplot(data=df, vars=['Date','Cases', 'Deaths','Recovered', 'Travel_history', 'Province','City'])
```

```python
# swarm plot
sns.swarmplot(data=df, y='Province', x='Cases')
```

```python
# barplot
sns.barplot(data=df, y='Cases', x='Province', ci=True)
plt.xticks(rotation=90)
```

```python
#scatter plot
sns.scatterplot(data=df,x='Cases', y='Deaths',hue='Province')
```

```python
# Pivot table (City, Cases)
table=pd.pivot_table(df, values= 'Cases', index='City', aggfunc=np.sum)
table
```

```python
# Pivot table (Province, Cases)
table=pd.pivot_table(df, values= 'Cases', index='Province', aggfunc=np.sum)
table
```

```python
# Correlation b/w Columns
correlation_matrix = df[['Cases', 'Deaths', 'Recovered']].corr()
sns.heatmap(correlation_matrix,annot=True)
```
