## 03 How to rename columns?

```python
#Without rename
df=pd.DataFrame({'Name':['Ali','Hassan','Abdullah','Ahmad'], 'Age':[17,18,19,20]})
df
```

```python
#with rename
df.rename(columns={'Name':'Name1','Age':'Age1'}, inplace=True)
df
```

### Another way to rename column

```python
df.columns=['Name_1','Age_1']
df
```

```python
#replace any character, string
df.columns=df.columns.str.replace('_',' ')
df
```

```python
df.columns=df.columns.str.replace(' ','_')
df
```

```python
#adding prefix to columns
df.add_prefix('First_')
```

```python
#adding suffix to columns
df.add_suffix(' |')
```

## 16- Splitting a string into multiple columns

```python
import pandas as pd
df=pd.DataFrame({'Name':["Rao Ali","Rao Hassan","Rao Abdullah","Rao Ahmad"],
              'Location':["Lodhran,Pakistan","Bahawalpur,Pakistan","Multan,Pakistan","Lahore,Pakistan"]})
df
```

```python
#splitting "Name" coloumn into two coloumns
df.Name.str.split(' ',expand=True)
```

```python
#Adding those splits into new coloumns
df[["first_name", "last_name"]]=df.Name.str.split(' ',expand=True)
df
```

```python
#Splitting "Location" coloumn into new columns
df.Location.str.split(',', expand=True)
```

```python
#Adding those columns into new columns
df[["City","Country"]]=df.Location.str.split(',',expand=True)
df
```

```python
#Refine data manipulation
df=df[["first_name","last_name","City","Country"]]
df
```

## 23- Reshaping a Dataframe

```python
#Creating Dataframe
fasla=pd.DataFrame([['12345',100,200,300],['34567',400,500,600],['67890',700,800,900]],
                    columns=['zip','factory','warehouse','retail'])
fasla.head()
```

```python
fasla.head().T
```

```python
#Creating new dataframe
fasla2=pd.DataFrame([[1,'12345','factory'],[2,'34567','warehouse']],
                    columns=['user_id','zip','location_type'])
fasla2.head()
```

```python
fasla
```

```python
#plot
sns.barplot(x='zip',y='factory',data=fasla)
```

```python
#convert long format
fasla_long=fasla.melt(id_vars='zip',var_name='location_type',value_name='distance')
fasla_long
```

```python
import seaborn as sns
sns.barplot(x='zip',y='distance', hue="location_type", data=fasla_long)
```
