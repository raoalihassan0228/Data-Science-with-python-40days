```python
import tensorflow as tf
import numpy as np 
from tensorflow import keras
```

```python
tf.__version__
```

```python
# simplets squential layer neural networks
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# compile the model (for correction of model)
model.compile(optimizer='sgd', loss = 'mean_squared_error')
```

```python
# make  numpy arrar for data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# training of neural netorks
model.fit(xs, ys, epochs=500)   #epochs mean that how many times loop will run
```

```python
# predict the model
print(model.predict([10]))
```

```python
print(model.predict([10, 20, 30]))
```
